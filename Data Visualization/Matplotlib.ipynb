# Matplotlib Histogram
- A histoghram is a graph showing frequency distribution.
- it is a graph showing number of obeservation in each given interval.

```python
import pandas as pd
```

```python
pd.set_option('display.max_columns',None)
```

```python
df=pd.read_csv('Automobile_data.csv')
df.head()
```

```python
df.shape[0]
```

```python
df.shape
```

```python
df.describe().T
```

```python
df.isna().sum()
```

```python
plt.hist(df['width'])
plt.title('Width frequency table')
plt.xlabel('Width')
plt.ylabel('Frequency')
```

```python
width_mean=df['width'].mean()
width_median=df['width'].median()

df['width'].hist(bins=10)
plt.axvline(width_mean,color='y')
plt.axvline(width_median,color='r')
plt.ylabel('Frequency')
plt.xlabel('Width')
```
