# Groupby function
- insted of calcutating avarege salary separatly,we can use pandas Groupby method

```python
#Groupby aggregate method
```

```python
df.groupby('rank')['salary'].mean()
```

```python
df.groupby('rank')['salary'].agg({'min', 'max','mean','median'})
```
